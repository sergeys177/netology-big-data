```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = pd.read_csv('house_prices.csv')
```

```python
str_columns = data.columns[data.dtypes == 'object']
```

```python
data = data.drop(str_columns, 1)
```

```python
float_columns = data.columns[data.dtypes == 'float64']
```

```python
data = data.drop(float_columns, 1)
```

```python
data.dtypes
```

```python
data.astype("int")
```

```python
y = data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(data,y,test_size=0.2)
```

```python
Xtrain.head()
```

```python
Ytrain.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(Xtrain, Ytrain)
```

```python
prediction = model.predict(Xtest)
```

```python
prediction[:5]
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(prediction, Ytest)
```
